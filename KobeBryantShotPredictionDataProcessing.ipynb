```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from sklearn import preprocessing
```

```python
data = pd.read_csv("data.csv")
```

```python
data_clean = data.dropna(subset = ["shot_made_flag"])
```

```python
# Make time_remaining variable instead of minutes and second remaining
data_clean.insert(len(data_clean.columns), "time_remaining", Series(data_clean.minutes_remaining*60 + data_clean.seconds_remaining))
```

```python
# Remove unnecessary columns
df = data_clean.drop('game_event_id', 1)
df = df.drop('game_id', 1)
df = df.drop('team_id', 1)
df = df.drop('team_name', 1)
df = df.drop('shot_id', 1)
df = df.drop('minutes_remaining', 1)
df = df.drop('seconds_remaining', 1)
df = df.drop('lat', 1)
df = df.drop('lon', 1)
df = df.drop('combined_shot_type', 1)
df = df.drop('loc_x', 1)
df = df.drop('loc_y', 1)
df = df.drop('season', 1)
```

```python
# Replace shot_type labels with numbers
shot_type_labels = df.shot_type.unique()
df.replace('2PT Field Goal', 0, inplace = True)
df.replace('3PT Field Goal', 1, inplace = True)
```

```python
# Replace shot_zone_area labels with numbers
shot_zone_area_labels = df.shot_zone_area.unique()
df.replace('Left Side(L)', 0, inplace = True)
df.replace('Left Side Center(LC)', 1, inplace = True)
df.replace('Right Side Center(RC)', 2, inplace = True)
df.replace('Center(C)', 3, inplace = True)
df.replace('Right Side(R)', 4, inplace = True)
df.replace('Back Court(BC)', 5, inplace = True)
```

```python
# Replace shot_zone_basic labels with numbers
shot_zone_basic_labels = df.shot_zone_basic.unique()
df.replace('Mid-Range', 0, inplace = True)
df.replace('Restricted Area', 1, inplace = True)
df.replace('In The Paint (Non-RA)', 2, inplace = True)
df.replace('Above the Break 3', 3, inplace = True)
df.replace('Right Corner 3', 4, inplace = True)
df.replace('Backcourt', 5, inplace = True)
df.replace('Left Corner 3', 6, inplace = True)
```

```python
# Replace shot_zone_range labels with numbers
shot_zone_range_labels = df.shot_zone_range.unique()
df.replace('8-16 ft.', 0, inplace = True)
df.replace('16-24 ft.', 1, inplace = True)
df.replace('Less Than 8 ft.', 2, inplace = True)
df.replace('24+ ft.', 3, inplace = True)
df.replace('Back Court Shot', 4, inplace = True)
```

```python
# Replace opponent labels with numbers
opponent_labels = df.opponent.unique()
df.replace('ATL', 0, inplace = True)
df.replace('BOS', 1, inplace = True)
df.replace('BKN', 2, inplace = True)
df.replace('CHA', 3, inplace = True)
df.replace('CHI', 4, inplace = True)
df.replace('CLE', 5, inplace = True)
df.replace('DAL', 6, inplace = True)
df.replace('DEN', 7, inplace = True)
df.replace('DET', 8, inplace = True)
df.replace('GSW', 9, inplace = True)
df.replace('HOU', 10, inplace = True)
df.replace('IND', 11, inplace = True)
df.replace('LAC', 12, inplace = True)
df.replace('LAL', 13, inplace = True)
df.replace('MEM', 14, inplace = True)
df.replace('MIA', 15, inplace = True)
df.replace('MIL', 16, inplace = True)
df.replace('MIN', 17, inplace = True)
df.replace('NJN', 18, inplace = True)
df.replace('NOH', 19, inplace = True)
df.replace('NOP', 20, inplace = True)
df.replace('NYK', 21, inplace = True)
df.replace('OKC', 22, inplace = True)
df.replace('ORL', 23, inplace = True)
df.replace('PHI', 24, inplace = True)
df.replace('PHX', 25, inplace = True)
df.replace('POR', 26, inplace = True)
df.replace('SAC', 27, inplace = True)
df.replace('SAS', 28, inplace = True)
df.replace('SEA', 29, inplace = True)
df.replace('TOR', 30, inplace = True)
df.replace('UTA', 31, inplace = True)
df.replace('WAS', 32, inplace = True)
df.replace('VAN', 33, inplace = True)
```

```python
# Convert shot made flag to int value
df['shot_made_flag'] = df['shot_made_flag'].astype(int)
```

```python
# Convert matchup to 0 for home game and 1 for away game
home = []
for x in df['matchup']:
    if 'vs.' in x:
        home.append(0)
    else:
        home.append(1)
df['home'] = home
df = df.drop('matchup', 1)
```

```python
# Convert game date to month and year columns
df.loc[:, 'month'] = df['game_date'].str[5:7].astype(int)
df.loc[:, 'year'] = df['game_date'].str[0:4].astype(int)
df = df.drop('game_date', 1)
```

```python
# Scale data
for col in df.columns:
    if col != 'shot_made_flag':
        df[col] = preprocessing.scale(df[col].values)
```

```python
# Make target column
target = df.shot_made_flag
df['target'] = target
df = df.drop('shot_made_flag', 1)
```

```python
df.to_csv('kobeshots.csv', index = False)
```
